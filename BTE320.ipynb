In Class Activity 1

```python
sales = float (input ("Enter sales amount ($)"))
cost_of_goods_sold = sales * .4
gross_profit = sales * .4
gross_profit = sales - cost_of_goods_sold
print(f" Total of goods sold: {cost_of_goods_sold}")
print(f" Gross profit: {gross_profit} $")
```

```python
# Prompt for tax rate
tax_rate = float(input("Enter tax rate (%): ")) / 100

# Calculate taxes and net income
taxes = gross_profit * tax_rate
net_income = gross_profit - taxes

# Display results
print(f"Taxes: {taxes}")
print(f"Net Income: {net_income}")
```

In Class Activity 2

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    result = num1 + num2
    print(result)

elif operator == "-":
    result = num1 - num2
    print(result)

elif operator == "*":
    result = num1 * num2
    print(result)

elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print(result)
    else:
        print("Error: Cannot divide by zero")

else:
    print("Invalid operator")
```

In Class Activity 4

```python
import math

ticketPrice = 10
initAttendees = 20

print("Ads\tProfit")
for ad in range(0, 201, 25):
    revenue = (initAttendees + 2 * round(math.sqrt(ad))) * ticketPrice
    costs = 200 + ad
    profit = revenue - costs
    print(f"{float(ad)}\t{float(profit)}")
```

In Class Activity 5

```python
import math

ticketPrice = 10
initAttendees = 20

print ("Ads/tProfit")

ad = 0
while ad < 201:
  revenue = (initAttendees + 2 * round(math.sqrt(ad))) * ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)}\t{float(profit)}")
  ad += 25
```

```python
print ("Hours/tFee")
hour = 1
while hour < 9:
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10
  elif fee > 20:
    fee = 20
  print(f"{float(hour)}\t{float(fee)}")
  hour += 1
```
