Income Statement using Simple Income Statement Calculation

```python
sales = float (input ("Enter sales amount ($)"))
cost_of_goods_sold = sales * .4
gross_profit = sales * .4
gross_profit = sales - cost_of_goods_sold
print(f" Total of goods sold: {cost_of_goods_sold}")
print(f" Gross profit: {gross_profit} $")
```

```python
# Prompt for tax rate
tax_rate = float(input("Enter tax rate (%): ")) / 100

# Calculate taxes and net income
taxes = gross_profit * tax_rate
net_income = gross_profit - taxes

# Display results
print(f"Taxes: {taxes}")
print(f"Net Income: {net_income}")
```

```python
import math

D = float(input("Enter annual demand (units per year): "))
S = float(input("Enter ordering cost ($ per order): "))
H = float(input("Enter holding cost ($ per unit per year): "))

EOQ = math.sqrt((2 * D * S) / H)

print("\nEconomic Order Quantity (EOQ):", round(EOQ, 2), "units per order")
```

Numerical Calculator

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    result = num1 + num2
    print(result)

elif operator == "-":
    result = num1 - num2
    print(result)

elif operator == "*":
    result = num1 * num2
    print(result)

elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print(result)
    else:
        print("Error: Cannot divide by zero")

else:
    print("Invalid operator")
```

Movie Theater (for/while loop)

For Loop

```python

print("Hours  Fee")

for hours in range(1, 9):
    fee = 5 + (2.5 * hours)
    fee = max(10, min(fee, 20))  # apply limits
    print(hours, round(fee, 1))
```

While loop

```python
print("Hours  Fee")

hours = 1

while hours <= 8:
    fee = 5 + (2.5 * hours)
    fee = max(10, min(fee, 20))  # apply limits
    print(hours, round(fee, 1))
    hours += 1
```

Rock-Paper-Scissors

```python
import random

check = input("Do you want to play a game (y/n)? ")

while check == 'y':
  player = input ("Choose b/w rock, paper, scissors: ")
  computer = random.choice(['rock', 'paper', 'scissors'])

  if player == computer:
      print('Draw')
  elif player == 'rock' and computer == 'scissors':
        print('Player wins!')
  elif player == 'paper' and computer == 'rock':
        print('Player wins!')
  elif player == 'scissors' and computer == 'paper':
        print('Player wins!')
  else:
      print('Computer wins!')

  check = input("Do you want to play again (y/n)? ")
```

Dynamic Dictionary; N distinct key:value pairs (while loop/if-elif-else statements)

```python
my_dict = {}
N = int(input("How many inputs do you want to enter? "))
count = 0
while count < N:
    key = input("Enter a name: ")

    if key not in my_dict:
        my_dict[key] = 1
    else:
        my_dict[key] += 1
    count += 1
print("\nFinal dictionary:")
print(my_dict)
```

ROI Calculator

```python
def roi(cost, revenue):
    assert cost != 0
    assert isinstance(cost, (int, float))
    assert isinstance(revenue, (int, float))
    print((revenue - cost) / cost)

roi(100, 150)
```

Loan Payment Calculator

```python
def loan_payment(principal, annual_rate, months):
    assert (isinstance(principal, (int, float)) and
            isinstance(annual_rate, (int, float)) and
            isinstance(months, int))
    if principal <= 0 or annual_rate <= 0 or months <= 0:
        raise ValueError("Invalid input")
    try:
        r = annual_rate / 12
        if r == 0:
            pmt = principal / months
        else:
            pmt = r * principal / (1 - (1 + r) ** (-months))
        return pmt
    except Exception as e:
        print(f"Error: {e}")
        return None

result = loan_payment(10000, 0.06, 60)
if result is not None:
    print(round(result, 2))
else:
    print("Loan payment calculation failed.")
```
